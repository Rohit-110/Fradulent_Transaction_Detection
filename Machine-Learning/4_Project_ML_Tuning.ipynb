# Model Building



## Train-Test Split

Split the dataset into train set and test set. Train set will be used to train the model. Test set will be used to check the performance of model

```python
# Split the y variable series and X variables dataset
X = df.drop(['isFraud'],axis=1)
y = df.isFraud.astype(bool)
```

```python
# Split the dataset into the training set and test set
from   sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.30, random_state = 42)
```

```python
print(X_train.shape,y_train.shape)
print(X_test.shape,y_test.shape)
```

## Model Fitting

```python
from sklearn.impute import KNNImputer, SimpleImputer

# replace inf
X_train = X_train.replace([np.inf, -np.inf], np.nan)
X_test = X_test.replace([np.inf, -np.inf], np.nan)

# Impute
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
# imputer = KNNImputer(n_neighbors=3)

X_train_imputed = imputer.fit_transform(X_train)
X_train_imputed = pd.DataFrame(X_train_imputed, columns=X_train.columns)
X_train_imputed.head()
```

```python
%%time
from lightgbm import LGBMClassifier
# Define the model
lgbc = LGBMClassifier(random_state=42, n_jobs = -1)

# Train the model
lgbc.fit(X_train,y_train)

lgbc
```

```python
# Prediction
y_pred_lgbc = lgbc.predict(X_test)
y_prob_pred_lgbc = lgbc.predict_proba(X_test)
y_prob_pred_lgbc = [x[1] for x in y_prob_pred_lgbc]
print("Y predicted : ",y_pred_lgbc)
print("Y probability predicted : ",y_prob_pred_lgbc[:5])
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score, roc_auc_score
from sklearn.metrics import ConfusionMatrixDisplay, RocCurveDisplay, PrecisionRecallDisplay
from sklearn.metrics import precision_recall_curve

# Make sure scikit-learn is up to date
!pip install -U scikit-learn
```

```python
def compute_evaluation_metric(model, x_test, y_actual, y_predicted, y_predicted_prob):
    # Accuracy score
    print("\n Accuracy Score : ", accuracy_score(y_actual, y_predicted))

    # AUC - ROC score
    print("\n AUC - ROC Score : ", roc_auc_score(y_actual, y_predicted_prob))

    # Classification report
    print("\n Classification Report : \n\n", classification_report(y_actual, y_predicted))

    # Confusion matrix
    cm = confusion_matrix(y_actual, y_predicted)
    print("\n Confusion Matrix : \n\n", cm)

    # Confusion matrix plot using ConfusionMatrixDisplay
    print("\n Confusion Matrix Plot: \n")
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=model.classes_)
    disp.plot(cmap='plasma')
    plt.show()

    # ROC curve plot using RocCurveDisplay
    print("\n AUC - ROC Curve:  \n")
    RocCurveDisplay.from_estimator(model, x_test,y_actual)
    plt.show()

    # Precision-Recall curve plot using PrecisionRecallDisplay
    print("\n Precision - Recall Curve:  \n")
    PrecisionRecallDisplay.from_estimator(model, x_test,y_actual)
    plt.show()

```

```python
# Compute Evaluation Metric
compute_evaluation_metric(lgbc, X_test, y_test, y_pred_lgbc, y_prob_pred_lgbc)
```

# Hyperparameter Tuning

**Hyperparameter** is a parameter that governs how the algorithm trains to learn the relationships. The values are set before the learning process begins.

**Hyperparameter tuning** refers to the automatic optimization of the hyper-parameters of a ML model.

## Model Fitting

```python
from sklearn.model_selection import RandomizedSearchCV, StratifiedKFold
from lightgbm import LGBMClassifier

# Define the estimator with balanced class weight
lgbmclassifier = LGBMClassifier(random_state=42)

# Define the parameters grid
param_grid = {
    'n_estimators': [100, 150, 200],  # Number of boosting iterations
    'num_leaves': [128, 256],  # Number of leaves in one tree
    'max_depth': [8, 10, 20],  # Maximum depth of tree
    'learning_rate': [0.05, 0.1],  # Step size shrinkage
    'scale_pos_weight': [len(y_train) / sum(y_train == 1)]  # Adjust weight for class imbalance
}

# Stratified cross-validation to maintain class distribution in folds
cv = StratifiedKFold(n_splits=3)

# Run randomized search with multiple cores
random_search = RandomizedSearchCV(
    lgbmclassifier,
    param_distributions=param_grid,
    n_iter=10,  # Number of combinations to sample
    refit=True,
    verbose=3,
    cv=cv,  # Use stratified K-fold
    n_jobs=-1,  # Use all available cores
    scoring='roc_auc'  # Use AUC for imbalanced datasets
)

# Fit the model for randomized search
random_search.fit(X_train_imputed, y_train)

# Display best parameters and best score
print("Best Parameters:", random_search.best_params_)
print("Best Score (AUC):", random_search.best_score_)
```

Get the best parameters corresponding to which you have best model

```python
# Since we used RandomizedSearchCV, replace 'grid' with 'random_search'
# (the variable used for the randomized search object).

# Best parameters after hyperparameter tuning
print("Best Parameters:", random_search.best_params_)

# Model parameters
print("Best Estimator:", random_search.best_estimator_)

# Use the best estimator
lgbmclassifier = random_search.best_estimator_
```

Let's use the best model to get predictions on test dataset. We would be looking at the predicted class and predicted probability both in order to evaluate the performance of the model

```python
# Impute X_Test before predicting
X_test_imputed = imputer.transform(X_test)

# Prediction using best parameters
y_grid_pred = lgbmclassifier.predict(X_test_imputed)
y_prob_grid_pred = lgbmclassifier.predict_proba(X_test_imputed)[:, 1]
print("Y predicted : ",y_grid_pred)
print("Y probability predicted : ",y_prob_grid_pred[:5])
```

## Model Evaluation

```python
# Compute Evaluation Metric
compute_evaluation_metric(lgbmclassifier, X_test_imputed, y_test, y_grid_pred, y_prob_grid_pred)
```
